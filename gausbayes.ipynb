```python
import math
data = [
    [45, 200, 130, 1, 1],
    [50, 180, 120, 3, 0],
    [38, 210, 135, 2, 1],
    [60, 220, 140, 1, 1],
    [42, 190, 125, 4, 0],
    [55, 215, 138, 1, 1],
    [47, 185, 128, 3, 0],
    [62, 230, 145, 1, 1],
    [40, 195, 130, 2, 0],
    [53, 205, 135, 1, 1],
    [36, 180, 120, 5, 0],
    [58, 225, 142, 1, 1],
    [44, 190, 125, 3, 0],
    [61, 218, 140, 1, 1],
    [39, 185, 123, 4, 0],
    [56, 220, 138, 1, 1],
    [41, 195, 128, 2, 0],
    [59, 230, 143, 1, 1],
    [37, 180, 122, 4, 0],
    [54, 210, 136, 1, 1]
]
stdata = data[:len(data)*4//5]
tdata = data[len(data)*4//5:]
def normbashxum(myu,d,x):
  return math.exp((-(x-myu)**2)/(2*d))/math.sqrt(2*d*math.pi)
def mijin(data):
  n = len(data)
  m = len(data[0])
  myu = []
  for j in range(m-1):
    s = 0
    for i in range(n):
      s+=data[i][j]
    myu.append(s/n)
  return myu
def dispersia(data,myu):
  n = len(data)
  m = len(data[0])
  d = []
  for j in range(m-1):
    s = 0
    for i in range(n):
      s+=(myu[j]-data[i][j])**2
    d.append(s/(n-1))
  return d
def bayes(data,x):
  target = []
  n = len(data)
  m = len(data[0])
  data0 = []
  data1 = []
  for i in range(n):
    target.append(data[i][-1])
    if data[i][-1] == 0:
      data0.append(data[i])
    else:
      data1.append(data[i])
  P00 = target.count(0)/n
  P01 = 1 - P00
  myu0 = mijin(data0)
  myu1 = mijin(data1)
  d0 = dispersia(data0,myu0)
  d1 = dispersia(data1,myu1)
  P0 = P00
  P1 = P01
  for j in range(m-1):
    P0 *= normbashxum(myu0[j],d0[j],x[j])
    P1 *= normbashxum(myu1[j],d1[j],x[j])
  if P0 > P1:
    return 0
  else:
    return 1

print("Testayin Data")
for i in tdata:
  print(i)
print("-------")
m = 0
print("Testayin Data Gushakum")
for i in tdata:
  y = bayes(stdata,i[:-1])
  print(i[:-1],y)
  m+=(y==i[-1])
print("-------")
print(f"Chshtutyun {m*100/len(tdata)}%")
print("-------")
age = int(input("Tariq :: "))
chol = int(input("Xolesterini makardaky :: "))
zarkchnshum = int(input("Zarkerakayin Chnshum :: "))
fiz = int(input("Fizikakan aktivutyuny :: "))
print("-------")
if bayes(stdata,[age,chol,zarkchnshum,fiz]) == 1:
  print("Katvaci risky mece")
else:
  print("Katvaci risky cacre")
```
