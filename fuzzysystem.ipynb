```python
import numpy as np

def triangular(x,a,b,c):
    if x <= a or x >= c:
        return 0,
    elif x > a and x < b:
        return (x - a) / (b - a)
    else:
        return (c - x) / (c - b)

def close(distance):
    return triangular(distance,0,0,20)

def medium(distance):
    return triangular(distance,15,25,40)

def far(d):
    return triangular(d,35,60,120)

def slow(speed):
    return triangular(speed, 0, 0, 40)

def moderate(speed):
    return triangular(speed, 30, 50, 70)

def fast(speed):
    return triangular(speed, 60, 80, 100)

def light(brake):
    return triangular(brake, 0, 0, 50)

def medium_brake(brake):
    return triangular(brake, 30, 50, 70)

def hard(brake):
    return triangular(brake, 60, 100, 100)

def fuzzify_distance(d):
    return {
        'close': close(d),
        'medium': medium(d),
        'far': far(d)
    }

def fuzzify_speed(v):
    return {
        'slow': slow(v),
        'moderate': moderate(v),
        'fast': fast(v)
    }

def apply_rules(distance_levels, speed_levels):

    light = 0
    medium = 0
    hard = 0

    hard = min(distance_levels['close'], speed_levels['fast'])
    medium = min(distance_levels['close'], speed_levels['moderate'])
    medium = max(medium, min(distance_levels['medium'], speed_levels['fast']))
    light = min(distance_levels['medium'], speed_levels['moderate'])
    light = max(light, distance_levels['far'])

    return {
        'light': light,
        'medium': medium,
        'hard': hard
    }

def defuzzify(output_levels):
    universe = np.linspace(0, 100, 500)
    aggregated = []

    for b in universe:
        mu_light = min(output_levels['light'], light(b))
        mu_medium = min(output_levels['medium'], medium_brake(b))
        mu_hard = min(output_levels['hard'], hard(b))
        aggregated.append(max(mu_light, mu_medium, mu_hard))

    numerator = sum(u * a for u, a in zip(universe, aggregated))
    denominator = sum(aggregated)

    return numerator / denominator if denominator != 0 else 0

def fuzzy_brake_controller(distance, speed):
    distance_levels = fuzzify_distance(distance)
    speed_levels = fuzzify_speed(speed)
    rule_outputs = apply_rules(distance_levels, speed_levels)
    brake_crisp = defuzzify(rule_outputs)
    return brake_crisp

distance = float(input("Enter the distance to the car in front (meters): "))
speed = float(input("Enter your current speed (km/h): "))

brake = fuzzy_brake_controller(distance, speed)

print(f"Brake intensity = {brake:.2f}%")
```
