```python
import numpy as np
import matplotlib.pyplot as plt
def parabol(x):
  return (x+4)**2
x1 = -6
x2= 4
x = np.linspace(x1, x2, 100)
a = 0.01
x0 = 3
xg = [x0]
plt.figure(figsize=(16, 8))
plt.xticks(np.arange(x1, x2+1, 0.5))
plt.plot(x, parabol(x))
while x1<=x0<=x2:
  x0 = x0 - a*2*(x0+4)
  xg.append(x0)
  if abs(a*2*(x0+4)) <= 0.00001:
    break
xg = np.array(xg)
plt.scatter(xg, parabol(xg),color='red')
plt.show()
```
