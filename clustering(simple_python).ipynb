```python
import math
import random
import copy
data = [
    [12, 8, 4, 4],
    [15, 7, 5, 5],
    [11, 8, 3, 3],
    [13, 7, 4, 4],
    [14, 8, 3, 5],
    [4, 6, 10, 1],
    [3, 5, 12, 0],
    [5, 6, 9, 1],
    [2, 5, 11, 0],
    [4, 6, 10, 1],
    [13, 7, 4, 4],
    [12, 8, 3, 4],
    [3, 5, 12, 0],
    [5, 6, 9, 1],
    [14, 8, 4, 5],
    [4, 6, 11, 1],
    [13, 7, 3, 4],
    [2, 5, 12, 0],
    [15, 8, 4, 5],
    [3, 6, 10, 1]
]
def evklid(x,y):
  n = len(x)
  d = 0
  for i in range(n):
    d += (x[i]-y[i])**2
  d = math.sqrt(d)
  return d
def cluster(data,k,miter):
  n = len(data)
  m = len(data[0])
  cpos = [[] for i in range(k)]

  for j in range(m):
    mn = 0
    mx = 0
    z = []
    for i in range(n):
      if data[i][j] > data[mx][j]:
        mx = i
        continue
      if data[i][j] < data[mn][j]:
        mn = i
    v = 0
    while v < k:
      c = random.randint(data[mn][j], data[mx][j])
      if c not in z:
        z.append(c)
        cpos[v].append(c)
        v+=1
  print(cpos)
  for iter in range(miter):
    clas = [[] for i in range(k)]
    d = []
    oldpos = copy.deepcopy(cpos)
    cpos = [[] for i in range(k)]
    for i in range(n):
      d.append([])
      for v in range(k):
        d[i].append(evklid(data[i],oldpos[v]))
      mn = 0
      for v in range(k):
        if d[i][mn]>d[i][v]:
          mn = v
      clas[mn].append(i)
    for j in range(m):
      for v in range(k):
          s = 0
          for i in clas[v]:
            s+=data[i][j]
          if len(clas[v]) != 0:
            cpos[v].append(s/len(clas[v]))
          else:
            cpos[v].append(oldpos[v][j])
    if cpos == oldpos:
      break
  print(cpos)
  print(clas)
cluster(data,2,1000)

```
